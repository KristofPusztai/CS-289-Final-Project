## Examining Frequency Regression for sine function:

We see that it is extremely hard to parameter fit a sine wave via gradient descent due to the multitude of local minima in the gradient of loss with respect to frequency for sine wave

```
import numpy as np
import matplotlib.pyplot as plt
```

```
X = np.array([i/100 for i in range(1000)])
Y = np.sin(2*X)
plt.plot(X, Y)
```

```
class sine_regression:
  def __init__(self, X, Y, iter = 20000, alpha = 0.1):
    self.X = X
    self.Y = Y
    self.a = 3
    self.alpha = alpha
    self.iter = iter
  def gradient(self, error):
    return np.mean(error*X)
  def regress(self):
    for i in range(self.iter):
      self.y_hat = np.sin(self.a * X)
      error = (self.y_hat - Y)
      self.a -= self.alpha * self.gradient(error)
    print(np.mean(error**2))
    print(self.a)
    return self.a
  def predict(self):
    return np.sin(self.a * X)
sreg = sine_regression(X, Y)
sreg.regress()
plt.plot(X, Y)
plt.plot(X, sreg.predict())
```

```
a_s = []
errors = []
for a in np.linspace(0,5,200):
  a_s.append(a)
  y_hat = np.sin(a * X)
  error = np.mean((Y - y_hat)**2)
  errors.append(error)
plt.plot(a_s, errors)
plt.xlabel('Frequency(a)')
plt.ylabel('MSE')
plt.show()
```
